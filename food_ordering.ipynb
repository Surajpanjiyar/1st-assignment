```python
class FoodItem:
    def __init__(self, food_id, name, quantity, price, discount, stock):
        self.food_id = food_id
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock

class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password

class Order:
    def __init__(self, user, items):
        self.user = user
        self.items = items
        self.total_amount = sum(item.price for item in items)

# Initialize data
food_items = [
    FoodItem(1, "Tandoori Chicken (4 pieces)", "4 pieces", 240, 0, 50),
    FoodItem(2, "Vegan Burger (1 Piece)", "1 piece", 320, 10, 30),
    FoodItem(3, "Truffle Cake (500gm)", "500gm", 900, 5, 20),
]

users = []
orders = []

# Function to add a new food item
def add_food_item(name, quantity, price, discount, stock):
    food_id = len(food_items) + 1
    food_item = FoodItem(food_id, name, quantity, price, discount, stock)
    food_items.append(food_item)
    print(f"Food item '{name}' added successfully!")

# Function to edit a food item by FoodID
def edit_food_item(food_id, name, quantity, price, discount, stock):
    for item in food_items:
        if item.food_id == food_id:
            item.name = name
            item.quantity = quantity
            item.price = price
            item.discount = discount
            item.stock = stock
            print(f"Food item '{name}' updated successfully!")
            return
    print(f"Food item with FoodID {food_id} not found.")

# Function to view the list of all food items
def view_food_items():
    print("List of Food Items:")
    for item in food_items:
        print(f"{item.food_id}. {item.name} [{item.quantity}] - INR {item.price} ({item.discount}% off) - Stock: {item.stock}")

# Function to remove a food item by FoodID
def remove_food_item(food_id):
    for item in food_items:
        if item.food_id == food_id:
            food_items.remove(item)
            print(f"Food item removed successfully!")
            return
    print(f"Food item with FoodID {food_id} not found.")

# Function to register a new user
def register_user(full_name, phone_number, email, address, password):
    user = User(full_name, phone_number, email, address, password)
    users.append(user)
    print(f"User '{full_name}' registered successfully!")

# Function to log in a user
def login_user(email, password):
    for user in users:
        if user.email == email and user.password == password:
            return user
    return None

# Function to place a new order
def place_order(user, selected_items):
    items = [food_items[item_index - 1] for item_index in selected_items]
    order = Order(user, items)
    orders.append(order)
    print("Order placed successfully!")
    print("Ordered Items:")
    for item in items:
        print(f"{item.name} - INR {item.price}")
    print(f"Total Amount: INR {order.total_amount}")

# Function to view order history
def view_order_history(user):
    print("Order History:")
    for order in orders:
        if order.user == user:
            print("Ordered Items:")
            for item in order.items:
                print(f"{item.name} - INR {item.price}")
            print(f"Total Amount: INR {order.total_amount}")

# Function to update user profile
def update_profile(user, full_name, phone_number, address, password):
    user.full_name = full_name
    user.phone_number = phone_number
    user.address = address
    user.password = password
    print("Profile updated successfully!")

# Main program
while True:
    print("\nWelcome to the Food Ordering App!")
    print("1. Admin Login")
    print("2. User Register")
    print("3. User Login")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        # Admin functionalities
        admin_password = input("Enter admin password: ")
        if admin_password == 'admin123':
            while True:
                print("\nAdmin Menu:")
                print("1. Add New Food Item")
                print("2. Edit Food Item")
                print("3. View Food Items")
                print("4. Remove Food Item")
                print("5. Logout")

                admin_choice = input("Enter your choice: ")

                if admin_choice == '1':
                    name = input("Enter food name: ")
                    quantity = input("Enter quantity (e.g., 100ml, 250gm, 4pieces): ")
                    price = float(input("Enter price: "))
                    discount = float(input("Enter discount percentage: "))
                    stock = int(input("Enter stock amount: "))
                    add_food_item(name, quantity, price, discount, stock)

                elif admin_choice == '2':
                    food_id = int(input("Enter FoodID to edit: "))
                    name = input("Enter new food name: ")
                    quantity = input("Enter new quantity (e.g., 100ml, 250gm, 4pieces): ")
                    price = float(input("Enter new price: "))
                    discount = float(input("Enter new discount percentage: "))
                    stock = int(input("Enter new stock amount: "))
                    edit_food_item(food_id, name, quantity, price, discount, stock)

                elif admin_choice == '3':
                    view_food_items()

                elif admin_choice == '4':
                    food_id = int(input("Enter FoodID to remove: "))
                    remove_food_item(food_id)

                elif admin_choice == '5':
                    break

                else:
                    print("Invalid choice. Please try again.")
        else:
            print("Admin password incorrect. Access denied.")

    elif choice == '2':
        # User registration
        full_name = input("Enter your full name: ")
        phone_number = input("Enter your phone number: ")
        email = input("Enter your email: ")
        address = input("Enter your address: ")
        password = input("Create a password: ")
        register_user(full_name, phone_number, email, address, password)

    elif choice == '3':
        # User login
        email = input("Enter your email: ")
        password = input("Enter your password: ")
        user = login_user(email, password)
        if user:
            while True:
                print("\nUser Menu:")
                print("1. Place New Order")
                print("2. Order History")
                print("3. Update Profile")
                print("4. Logout")

                user_choice = input("Enter your choice: ")

                if user_choice == '1':
                    print("Available Food Items:")
                    view_food_items()
                    selected_items = input("Enter food item numbers (e.g., 1 2 3): ").split()
                    selected_items = [int(item) for item in selected_items]
                    place_order(user, selected_items)

                elif user_choice == '2':
                    view_order_history(user)

                elif user_choice == '3':
                    full_name = input("Enter new full name: ")
                    phone_number = input("Enter new phone number: ")
                    address = input("Enter new address: ")
                    password = input("Enter new password: ")
                    update_profile(user, full_name, phone_number, address, password)

                elif user_choice == '4':
                    break

                else:
                    print("Invalid choice. Please try again.")
        else:
            print("Invalid login credentials. Please try again.")

    elif choice == '4':
        print("Thank you for using the Food Ordering App. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```
